# Multilayer Perceptron com Keras

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.utils import np_utils
import matplotlib.pyplot as plt
from keras.layers import Dense
import pandas as pd
```

```python
base = pd.read_csv('iris.csv')
entradas = base.iloc[:, 0:4].values 
classes = base.iloc[:,4].values
```

Transformação das saídas - categórico para numérico

```python
labelEncoder = LabelEncoder()
classes = labelEncoder.fit_transform(classes)
classes_dummy = np_utils.to_categorical(classes)
print(classes_dummy)
```

### Criação das bases de treinamento e teste

25% para testes 

```python
entradas_trei, entradas_test, classes_trei, classes_test = train_test_split(entradas, classes_dummy, test_size=0.25)
```

Estrutura da Rede Neural

```python
rede = Sequential()
# 4 neurônios na camada escondida
# tamanho da entrada = 4
rede.add(Dense(units = 4, activation="relu", input_dim = 4))
# mais uma camada escondida
rede.add(Dense(units = 4, activation= "relu"))
# camada de saída
rede.add(Dense(units = 3, activation="softmax"))
# compilar a rede neural
rede.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["categorical_accuracy"])
```

![Sample Dog Output](../imagens/1mlpKeras.jpg)

### Treinamento da Rede Neural

```python
historico = rede.fit(entradas_trei, classes_trei, batch_size=10, epochs=1000)
```

### Avaliação da rede

```python
resultado = rede.evaluate(entradas_test, classes_test)
```

Perca e acurácia

```python
print(resultado)
```

# Matriz de confusão

### Organização das saídas

```python
previsoes = rede.predict(entradas_test)
previsoes = (previsoes > 0.5)
import numpy as np
classes_test2 = [np.argmax(t) for t in classes_test]
print(classes_test2)
previsoes2 =  [np.argmax(t) for t in previsoes]
print(previsoes2)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import itertools
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Matriz de Confusão',
                          cmap=plt.cm.Blues):   

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Saída correta')
    plt.xlabel('Saída encontrada')    
```

```python
cm = confusion_matrix(classes_test2, previsoes2)
plot_confusion_matrix(cm, 
                      classes=['setosa', 'virginica', 'versicolor'],
                      title='Matriz de Confusão - Iris Flower')
```
